## 2. Exploring Relationships

The interactive graph above lets you pan, zoom, and click nodes to inspect their properties (risk score, mitigation status, regulatory framework). The cells below generate **focused subgraphs** that slice the data by dimension — directly supporting the Component 3 analysis in Slides 13–16.

Interesting relationships to explore:

- All risks in a given **category** (e.g. Human Factor risks — the largest cluster)
- All risks governed by a specific **regulation** (e.g. EU AI Act Art. 14 Human Oversight)
- All risks **affecting a given stakeholder** (e.g. SOC Analyst risk exposure)
- Most **connected nodes** (degree centrality) — regulatory pressure points


### 2.1 Risks in a Category

Filter the graph to a single risk category. Change `target_category` to any of:
`Cyber`, `Privacy`, `Accountability`, `Human Factor`, `Transparency`, `Compliance`.

The example below shows **Human Factor** risks — the category carrying the highest average risk score (16.0) because it contains R11 (automation bias, CRITICAL) and R13 (alert fatigue, CRITICAL), both directly engaging EU AI Act Art. 14.


```python
# Subgraph: risks in a single category + their regulations and stakeholders
target_category = "Human Factor"

cat_risk_ids = [
    n["id"] for n in w.nodes
    if n["properties"].get("type") == "Risk"
    and n["properties"].get("category") == target_category
]

# Include the category hub and all first-order neighbours of those risks
keep_ids = set(cat_risk_ids)
keep_ids.add("CAT-" + {
    "Cyber": "CYBER", "Privacy": "PRIVACY", "Accountability": "ACCOUNT",
    "Human Factor": "HUMAN", "Transparency": "TRANSP", "Compliance": "COMPLY"
}[target_category])

for e in w.edges:
    if e["start"] in cat_risk_ids:
        keep_ids.add(e["end"])
    if e["end"] in cat_risk_ids:
        keep_ids.add(e["start"])

cat_nodes = [n for n in w.nodes if n["id"] in keep_ids]
cat_edges = [e for e in w.edges if e["start"] in keep_ids and e["end"] in keep_ids]

cat_graph = GraphWidget()
cat_graph.nodes = cat_nodes
cat_graph.edges = cat_edges
cat_graph.directed = True
cat_graph.set_node_color_mapping(custom_node_color_mapping)

print(f"Category: {target_category}")
print(f"  Risks in this category: {len(cat_risk_ids)}")
print(f"  Nodes in subgraph:      {len(cat_nodes)}")
print(f"  Edges in subgraph:      {len(cat_edges)}")
display(cat_graph)
```

### 2.2 Risks Governed by a Specific Regulation

Filter the graph to show which risks are constrained by a given regulatory obligation. This surfaces **cross-framework tensions** (e.g. the same risk governed by both the EU AI Act and GDPR).

The example targets **EU AI Act Article 14 — Human Oversight**, which the Reflective Commentary (Slide 17) identifies as the regulatory linchpin for the human-centred analysis.


```python
# Subgraph: all risks GOVERNED_BY a specific regulatory node
target_regulation = "EUAI-14"  # EU AI Act Art. 14 — Human Oversight

# Find risks pointing to this regulation with a GOVERNED_BY edge
governed_risks = [
    e["start"] for e in w.edges
    if e["end"] == target_regulation
    and e["properties"].get("label") == "GOVERNED_BY"
]

# Build subgraph: the regulation + governed risks + their categories + their stakeholders
keep_ids = set(governed_risks) | {target_regulation}
for e in w.edges:
    if e["start"] in governed_risks and e["properties"].get("label") in ("CATEGORISED_AS", "AFFECTS"):
        keep_ids.add(e["end"])

reg_nodes = [n for n in w.nodes if n["id"] in keep_ids]
reg_edges = [e for e in w.edges if e["start"] in keep_ids and e["end"] in keep_ids]

reg_graph = GraphWidget()
reg_graph.nodes = reg_nodes
reg_graph.edges = reg_edges
reg_graph.directed = True
reg_graph.set_node_color_mapping(custom_node_color_mapping)

target_label = next((n["properties"]["label"] for n in w.nodes if n["id"] == target_regulation), target_regulation)
print(f"Regulation: {target_label}")
print(f"  Risks governed by this article: {len(governed_risks)}  -> {governed_risks}")
print(f"  Nodes in subgraph:              {len(reg_nodes)}")
print(f"  Edges in subgraph:              {len(reg_edges)}")
display(reg_graph)
```

Change `target_stakeholder` to any of: `STK-SOC`, `STK-CISO`, `STK-DPO`, `STK-VENDOR`, `STK-EMP`, `STK-BOARD`.


```python
# Subgraph: all risks affecting a single stakeholder
target_stakeholder = "STK-SOC"  # SOC Analyst

affecting_risks = [
    e["start"] for e in w.edges
    if e["end"] == target_stakeholder
    and e["properties"].get("label") == "AFFECTS"
]

# Include risks + stakeholder + the regulations those risks are governed by
keep_ids = set(affecting_risks) | {target_stakeholder}
for e in w.edges:
    if e["start"] in affecting_risks and e["properties"].get("label") == "GOVERNED_BY":
        keep_ids.add(e["end"])

stk_nodes = [n for n in w.nodes if n["id"] in keep_ids]
stk_edges = [e for e in w.edges if e["start"] in keep_ids and e["end"] in keep_ids]

stk_graph = GraphWidget()
stk_graph.nodes = stk_nodes
stk_graph.edges = stk_edges
stk_graph.directed = True
stk_graph.set_node_color_mapping(custom_node_color_mapping)

# Compute average risk score for this stakeholder
scores = [
    n["properties"]["score"] for n in w.nodes
    if n["id"] in affecting_risks and n["properties"].get("type") == "Risk"
]
avg_score = sum(scores) / len(scores) if scores else 0

target_label = next((n["properties"]["label"] for n in w.nodes if n["id"] == target_stakeholder), target_stakeholder)
print(f"Stakeholder: {target_label}")
print(f"  Risks affecting this stakeholder: {len(affecting_risks)}  -> {affecting_risks}")
print(f"  Average risk score:               {avg_score:.1f} / 25")
print(f"  Nodes in subgraph:                {len(stk_nodes)}")
print(f"  Edges in subgraph:                {len(stk_edges)}")
display(stk_graph)
```

### 2.4 High Degree Centrality — Regulatory Pressure Points

Nodes with a high number of incoming/outgoing edges are **regulatory pressure points** — obligations or stakeholders that sit at the intersection of many risks. In a real deployment, these are the items that most deserve early investment: mitigating one high-centrality node indirectly reduces exposure across many risks.


```python
# Calculate degree centrality for every node (count of incident edges)
degree_centrality = {}
for node in w.nodes:
    degree_centrality[node["id"]] = sum(
        1 for edge in w.edges
        if edge["start"] == node["id"] or edge["end"] == node["id"]
    )

# Print the top 10 most-connected nodes
print("Top 10 nodes by degree centrality (regulatory pressure points):")
print("-" * 70)
for nid, deg in sorted(degree_centrality.items(), key=lambda x: -x[1])[:10]:
    lbl = next((n["properties"]["label"] for n in w.nodes if n["id"] == nid), nid)
    print(f"  {deg:>3}  {nid:<10}  {lbl}")

# Subgraph of nodes above a threshold
threshold = 5
high_degree_ids = [nid for nid, deg in degree_centrality.items() if deg > threshold]
high_degree_nodes = [n for n in w.nodes if n["id"] in high_degree_ids]
high_degree_edges = [
    e for e in w.edges
    if e["start"] in high_degree_ids and e["end"] in high_degree_ids
]

high_degree_graph = GraphWidget()
high_degree_graph.nodes = high_degree_nodes
high_degree_graph.edges = high_degree_edges
high_degree_graph.directed = True
high_degree_graph.set_node_color_mapping(custom_node_color_mapping)

print(f"\nSubgraph (degree > {threshold}): {len(high_degree_nodes)} nodes, {len(high_degree_edges)} edges")
display(high_degree_graph)
```

## 3. Risk Register Summary Statistics

Reproduces the automated summary shown on **Slide 20** (Risk-Compliance Mapping Dashboard) directly from the graph data. This provides audit evidence that the graph, Excel template, and slide deck share a single source of truth.


```python
# Compute risk register summary from the graph nodes
risks = [n for n in w.nodes if n["properties"].get("type") == "Risk"]

# Severity distribution
level_counts = {"CRITICAL": 0, "HIGH": 0, "MEDIUM": 0, "LOW": 0}
for r in risks:
    level_counts[r["properties"]["level"]] += 1

# Mitigation status distribution
status_counts = {}
for r in risks:
    s = r["properties"]["mitigation_status"]
    status_counts[s] = status_counts.get(s, 0) + 1

# Average score
avg_score = sum(r["properties"]["score"] for r in risks) / len(risks)

# Unmitigated gaps
gaps = [r for r in risks if r["properties"]["mitigation_status"] == "No Mitigation"]

# Top critical risks
top_critical = sorted(
    [r for r in risks if r["properties"]["level"] == "CRITICAL"],
    key=lambda r: -r["properties"]["score"]
)

print("=" * 70)
print("AI-DRIVEN SIEM/SOAR — RISK REGISTER SUMMARY")
print("=" * 70)
print(f"Total risks: {len(risks)}")
print(f"Average risk score: {avg_score:.1f} / 25")
print()
print("Severity distribution:")
for level, count in level_counts.items():
    bar = "#" * count
    print(f"  {level:<9} {count:>3}  {bar}")
print()
print("Mitigation status:")
for status, count in sorted(status_counts.items(), key=lambda x: -x[1]):
    print(f"  {status:<15} {count}")
print()
print(f"Unmitigated gaps ({len(gaps)}):")
for r in gaps:
    print(f"  {r['id']}  (score {r['properties']['score']}, {r['properties']['level']})  {r['properties']['label']}")
print()
print("Top CRITICAL risks:")
for r in top_critical:
    print(f"  {r['id']}  score {r['properties']['score']}  {r['properties']['label']}")
```
